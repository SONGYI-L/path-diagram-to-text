```python
import os

# the path to the folder containing the images
folder_path = "/content/drive/MyDrive/graph-ocr"

# get the names of all the files in the folder
file_names = os.listdir(folder_path)
```

```python
# create a new list to store the paths to the images
image_paths = []

# loop through the file names and create the full paths
for file_name in file_names:
    # get the full path of the file
    file_path = os.path.join(folder_path, file_name)

    # append the file path to the list of image paths
    image_paths.append(file_path)
```

```python
# open a new file for writing
with open("/content/drive/MyDrive/graph-ocr/test.txt", "w") as f:
    # loop through the image paths and write them to the file
    for image_path in image_paths:
        f.write(image_path + "\n")
```

```python
with open('/content/drive/MyDrive/graph-ocr/test.txt', 'r') as f:
    test_img_path=[]
    for line in f:
        test_img_path.append(line.strip())
print(test_img_path)
```

```python
import paddlehub as hub

# 加载移动端预训练模型
#ocr = hub.Module(name="chinese_ocr_db_crnn_mobile")
# 服务端可以加载大模型，效果更好 
ocr = hub.Module(name="chinese_ocr_db_crnn_server") 
```

```python
import cv2
import json
import os

# 读取测试文件夹test.txt中的照片路径
np_images =[cv2.imread(image_path) for image_path in test_img_path] 

# create a list of image names from the image paths
img_names = [os.path.basename(path) for path in test_img_path]

results = ocr.recognize_text(
                    images=np_images,         # 图片数据，ndarray.shape 为 [H, W, C]，BGR格式；
                    use_gpu=False,            # 是否使用 GPU；若使用GPU，请先设置CUDA_VISIBLE_DEVICES环境变量
                    output_dir='/content/drive/MyDrive/ocr_result',  # 图片的保存路径，默认设为 ocr_result；
                    visualization=False,       # 是否将识别结果保存为图片文件；
                    box_thresh=0.5,           # 检测文本框置信度的阈值；
                    text_thresh=0.5)          # 识别中文文本置信度的阈值；
result_list = []

for result, name in zip(results, img_names):
    data = result['data']
    save_path = result['save_path']
    image_info = {'save_path': name, 'data': []}
    for infomation in data:
        # append the text, confidence, and text box position to the dictionary
        image_info['data'].append({
            'text': infomation['text'],
            'confidence': infomation['confidence'],
            'text_box_position': infomation['text_box_position']
        })
    result_list.append({'name': name, 'data': image_info})

# print the result
print(result_list)
# convert the list to json string
json_str = json.dumps(result_list)

# write the json string to a file
with open('/content/drive/MyDrive/output.json', 'w') as f:
    f.write(json_str)
```
